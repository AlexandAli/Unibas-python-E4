# Import the required modules.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Age plots


### Use pandas to read the excel file.

```python
df = pd.read_csv('fake-age-data.csv')

df.head()
```

Let's calculate how full decades (10 years) they have lived.

```python
df_decade = df/10
df_decade = df_decade.astype(int)
# Important to remember that astype(int) does not round it just chops off the decimals!!!

df_decade.head()
```

```python
print("The length of the dataset is "+str(len(df))+".")
print("The mean of the dataset is "+str(round(df['age'].mean(), 2))+".")
df.describe()
```

Note: converting to an integer does not round! It merely drops off the decimal values.

### Create the desired plots.

```python
sns.set_context('talk')
sns.distplot(df, bins = 10)
plt.xlabel('Age')
plt.ylabel('Normalized frequency')
plt.show()
plt.close()
```

### Which statistical metric is easy to see with this plot?
Mean, median, mode, standard deviation, or standard error

```python
fig, ax =plt.subplots(figsize = (3, 4.5))
sns.barplot(data = df, color = 'gold')
ax.set_xticklabels(labels=['Participants'])
plt.ylabel('Age')
plt.show()
plt.close()
```

```python
fig, ax =plt.subplots(figsize = (3, 4.5))
sns.violinplot(data = df, color = 'gold')
ax.set_xticklabels(labels=['Participants'])
plt.ylabel('Age')
plt.show()
plt.close()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

ax1 = sns.distplot(df, bins = 10, ax=axs[0,0])
ax1.set_xlabel('Age')
ax1.set_ylabel('Normalized Frquency')

ax2 = sns.barplot(data = df, ax=axs[0, 1])
ax2.set_ylabel('Age')
ax2.set_xticklabels(labels=['Participants'])

ax3 = sns.boxplot(data = df, ax=axs[1, 0])
ax3.set_ylabel('Age')
ax3.set_xticklabels(labels=['Participants'])

ax4 = sns.violinplot(data = df, ax=axs[1, 1])
ax4.set_ylabel('Age')
ax4.set_xticklabels(labels=['Participants'])

plt.subplots_adjust(wspace= 0.3, hspace = 0.3)

plt.show()
plt.close()
```

## Let's look at some hypothetical data regarding catalysis.

Import the data.

```python
df_cat = pd.read_excel('fake-catalysis-data.xlsx')

print("The dataset length is "+ str(len(df_cat))+".")
df_cat.head()
```

The dataset has two columns: i) the day the data was collected and ii) the total turnover number (TTN).
We can look at the statistical analysis with pandas.

```python
df_cat.describe()
```

We can plot the data to visually examine the data.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

ax1 = sns.distplot(df_cat['TTN'], ax=axs[0,0])
ax1.set_xlabel('TTN')
ax1.set_ylabel('Normalized Frquency')

ax2 = sns.barplot(data = df_cat['TTN'], ax=axs[0, 1])
ax2.set_ylabel('TTN')
ax2.set_xticklabels(labels=['Catalyst 1'])

ax3 = sns.boxplot(data = df_cat['TTN'], ax=axs[1, 0])
ax3.set_ylabel('TTN')
ax3.set_xticklabels(labels=['Catalyst 1'])

ax4 = sns.violinplot(data = df_cat['TTN'], ax=axs[1, 1])
ax4.set_ylabel('TTN')
ax4.set_xticklabels(labels=['Catalyst 1'])

plt.subplots_adjust(wspace= 0.4, hspace = 0.3)

plt.show()
plt.close()
```

```python
fig, ax =plt.subplots(figsize = (8, 4.5))

sns.violinplot(data = df_cat, y = df_cat['TTN'], x=df_cat['Day'])

sns.swarmplot(data = df_cat, y = df_cat['TTN'], x=df_cat['Day'], color='white', size = 6)

plt.show()
plt.close()
```

```python
sns.scatterplot(data = df_cat, x = [0]*len(df_cat), y = df_cat['TTN'],
               hue = df_cat['Day'], legend= False)
plt.show()
plt.close()
```

Note: in the scatterplot above, x is artificially set to zero because we have only one column. If there were multiple columns, we would set x as a function of the column number.
